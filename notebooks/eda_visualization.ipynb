```python
import os
import pandas as pd

# Define the folder path
folder_path = r'D:\DATA KERJA\Project GITHUB\Python for Data Science Exploratory Data Analysis and Visualization\data'

# List all files in the folder
all_files = os.listdir(folder_path)

# Filter for CSV files
csv_files = [f for f in all_files if f.endswith('.csv')]

# Create an empty dictionary to store DataFrames
dataframes = {}

# Loop through CSV files and read them into DataFrames
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dataframes[file] = df  # Store DataFrame in dictionary with the filename as key

# Display the first few rows of each DataFrame
for file, df in dataframes.items():
    print(f"First few rows of {file}:")
    print(df.head(), "\n")
```

```python
# Check for missing values
print(df.info())

# Summary statistics for numerical columns
print(df.describe())

# Summary statistics for categorical columns
print(df.describe(include=['object']))
```

```python
# Check for missing values
print(df.isnull().sum())

# Handling missing values (if any)
df.fillna(method='ffill', inplace=True)  # Example: Forward fill
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Fill missing numerical values with mean
df['avg_rating'].fillna(df['avg_rating'].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of each numerical feature
df.hist(figsize=(10, 8))
plt.show()

# Boxplot for each numerical feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.show()
```

```python
# Convert date columns to datetime
df['created'] = pd.to_datetime(df['created'])
df['published_time'] = pd.to_datetime(df['published_time'])
```

```python
# Calculate discount percentage
df['discount_percentage'] = ((df['price_detail__amount'] - df['discount_price__amount']) / df['price_detail__amount']) * 100

# Fill missing discount values with 0 without using inplace
df['discount_percentage'] = df['discount_percentage'].fillna(0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for numerical features
df[['num_subscribers', 'avg_rating', 'num_reviews']].hist(bins=15, figsize=(15, 6), layout=(2, 3))
plt.show()

# Boxplot for price details
sns.boxplot(x='price_detail__currency', y='price_detail__amount', data=df)
plt.title('Price Distribution by Currency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only the numerical columns for the pair plot
numerical_columns = ['num_subscribers', 'avg_rating', 'avg_rating_recent', 'num_reviews', 'discount_percentage']
sns.pairplot(df[numerical_columns])
plt.show()

# Select only the numerical columns for the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
